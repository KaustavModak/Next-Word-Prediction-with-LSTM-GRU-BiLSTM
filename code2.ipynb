```python
import numpy as np
import pandas as pd
```

```python
# Load the Dataset
with open('hamlet.txt','r') as file:
    data=file.read().lower() # lowering the alphabets
```

```python
print(data)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts([data])
```

```python
tokenizer.word_counts
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index) # vocabulary count
```

```python
for sentence in data.split("\n"):
    print(sentence)
```

```python
for sentence in data.split("\n"):
    print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
input_sequences=[]
for sentence in data.split("\n"):
    tokenized_text=tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1,len(tokenized_text)):
        input_sequences.append(tokenized_text[:i+1])

input_sequences
```

```python
max_len=max(len(x) for x in input_sequences)
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_text=pad_sequences(input_sequences,maxlen=max_len,padding='pre')
padded_text
```

```python
X=padded_text[:,:-1]
y=padded_text[:,-1]
```

```python
X
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
len(tokenizer.word_index) #vocab count
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=4818)
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_Test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Embedding,LSTM,Dropout,GRU,Bidirectional
```

```python
model=Sequential()
model.add(Embedding(4818,100,input_length=max_len-1))
model.add(LSTM(150,return_sequences=True))
model.add(Dropout(0.2)) # rate=0.2 --> freq in which i/ps are dropped to prevent overfitting
model.add(LSTM(100))
model.add(Dense(4818,activation="softmax"))
```

```python
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=['accuracy'])
model.summary()
```

```python
model.input_shape
```

```python
history=model.fit(X_train,y_train,epochs=100,validation_data=[X_Test,y_test])
```

```python
text="You come"

for i in range(10):
    token_text=tokenizer.texts_to_sequences([text])[0]
    padded_text=pad_sequences([token_text],maxlen=max_len-1,padding='pre')
    pos=np.argmax(model.predict(padded_text))
    for word,index in tokenizer.word_index.items():
        if index==pos:
            text=text+" "+word
            print(text)
```
